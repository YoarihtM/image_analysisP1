```python

import numpy as np
import math
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
"""Código para Histograma"""

def histograma(pic_arr):
  if pic_arr.ndim==3:
    hist1 = cv2.calcHist([pic_arr], [0], None, [255], [0, 255])
    hist2 = cv2.calcHist([pic_arr], [1], None, [255], [0, 255])
    hist3 = cv2.calcHist([pic_arr], [2], None, [255], [0, 255])
    plt.plot(hist1, color='red')
    plt.plot(hist2, color='green')
    plt.plot(hist3, color='blue')
  else:
    hist = cv2.calcHist([pic_arr], [0], None, [255], [0, 255])
    plt.plot(hist, color='gray')
  plt.xlabel('Intensidad de iluminacion')
  plt.ylabel('Cantidad de pixeles')
  plt.show()
```

```python
"""Código para Desplazamiento"""

def desplazamiento(pic_arr,des):
  pic_arr2 = pic_arr.copy()
  height = pic_arr2.shape[0]
  width = pic_arr2.shape[1]
  if pic_arr2.ndim == 3:
    for channel in range(0,3):   #PARA LOS CANALES (R,G,B)
      for i in range(0,height):   
        for j in range(0,width):  
            if pic_arr2[i,j,channel] + des >255:
              pic_arr2[i,j,channel] = 255
            if pic_arr2[i,j,channel] + des <0:
              pic_arr2[i,j,channel] = 0
            else:
              pic_arr2[i,j,channel] = pic_arr2[i,j,channel] + des
  else:  
    for i in range(0,height):   
      for j in range(0,width):   
        if pic_arr2[i,j] + des > 255:
          pic_arr2[i,j] = 255
        if pic_arr2[i,j] + des < 0:
          pic_arr2[i,j] = 0
        else:
          pic_arr2[i,j] = pic_arr2[i,j] + des
  return pic_arr2
```

```python
"""Código para Contracción"""

def contraccion(pic_arr, Cmax, Cmin):
  pic_result = pic_arr.copy()
  arr_MAX = np.amax(pic_arr)
  arr_MIN = np.amin(pic_arr)
  height = pic_arr.shape[0]
  widthh = pic_arr.shape[1]
  if True:
    #for channel in range(0,3):   #PARA LOS CANALES (R,G,B)
    for i in range(0,height):   
      for j in range(0,widthh):
        aux = ((((Cmax - Cmin)/ (arr_MAX - arr_MIN)) * (pic_result[i,j]-arr_MIN)) + Cmin)
        aux2 = np.round(aux,0)  
        pic_result[i,j] = aux2
  else:
    for i in range(0,height):   
      for j in range(0,widthh): 
        aux = ((((Cmax - Cmin)/ (arr_MAX - arr_MIN)) * (pic_arr[i,j]-arr_MIN)) + Cmin)
        pic_result[i,j] = round(aux)

    
  return pic_result
```

```python
"""Código para Expansión"""

def expansion(pic_arr, MAX, MIN):
   pic_result = pic_arr.copy()
   arr_MAX = np.amax(pic_arr)#VALOR MAXIMO
   arr_MIN = np.amin(pic_arr)#VALOR MINIMO
   height = pic_arr.shape[0]
   width = pic_arr.shape[1]
   if True:
    #for channel in range(0,3):   #PARA LOS CANALES (R,G,B)
    for i in range(0,height):   
      for j in range(0,width):
        aux = ((((pic_arr[i,j] - arr_MIN)/(arr_MAX - arr_MIN) ) * (MAX - MIN) ) + MIN)
        aux2 = np.round(aux,0)  
        pic_result[i,j] = aux2
   else:
    for i in range(0,height):   
      for j in range(0,width):  
       aux = ((((pic_arr[i,j] - arr_MIN)/(arr_MAX - arr_MIN) ) * (MAX - MIN)) + MIN)
       pic_result[i,j] = round(aux)

   return pic_result
```

```python
"""Código para Ecualización Rayleigh"""

# Alfa permite variar el crecimiento de la distribucion

def equalization(pic_arr, value):
   pic_result = pic_arr.copy()
   arr_MAX = np.amax(pic_arr)#VALOR MAXIMO
   print("valor maximo", arr_MAX)
   arr_MIN = np.amin(pic_arr)#VALOR MINIMO
   print("valor minimo", arr_MIN)
   height = pic_arr.shape[0]
   width = pic_arr.shape[1]
   MAX=value
   MIN=1
   tamanio = len(pic_arr)
   print("tamanio", tamanio)
   MN = tamanio * tamanio
   #numero de apariciones de cada valor
   repeticiones = np.unique(pic_arr, return_counts=True)
   nk = repeticiones[1]
   print("NK=",nk)
   rk = repeticiones[0]
   print("RK=",rk)
   pRrK = []
   aux1 = []
   pRrK = nk/MN
   aux1 = pRrK
   alfa = value
   alfaCuadrado = pow(alfa,2)
   print(alfaCuadrado)

   # Almacenamos el valor Pr(Rk) en un array para facilitar el uso de los resultados
   for r in range(len(pRrK)):
    aux1[r] = arr_MIN + (math.sqrt(((2*alfaCuadrado)*(math.log(1/(1-pRrK[r]))))))
    #aux1[r] = arr_MIN + (pow(((2*alfaCuadrado)*(math.log(1/(1-pRrK[r])))),0.5))
    
   print(aux1) 
    

   if pic_arr.ndim == 3:
    for channel in range(0,3):   #PARA LOS CANALES (R,G,B)
      for i in range(0,height):   
        for j in range(0,width):
          aux = pic_arr[i,j, channel]
          #print("Aux en array original:",aux)
          if aux > 0:
            for m in range(len(aux1)):
                if aux == rk[m]:
                    aux = aux1[m]
            #print("aux final=", aux)
          else:
            aux = 1
          aux2 = np.round(aux,0)  
          pic_result[i,j,channel] = aux2
   else:
    for i in range(0,height):   
      for j in range(0,width):
        aux = pic_arr[i,j]
        #print("aux=", aux)
        if aux>0:
            for m in range(len(aux1)):
                if aux == rk[m]:
                    aux = aux1[m]
            #print("aux final=", aux)
        else:
          aux = 1
        pic_result[i,j] = round(aux)
        
   return pic_result
```

```python
"""## **PRUEBA 1**"""

#IMAGEN ORIGINAL Y EJECUCIÓN DE FUNCIONES
filename="Imagen5.bmp"
prueba1=Image.open(filename) #Abrimos la imagen 1
prueba1_arr = np.asarray(prueba1) #Convierte imagen a arreglo

"""#### Original v. Desplazamiento
*   Valor de desplazamiento = 50
"""

prueba1_desplazada = desplazamiento(prueba1_arr,70) #Desplazamiento
img_prueba1_desplazada = Image.fromarray(prueba1_desplazada)
histograma(prueba1_arr)
histograma(prueba1_desplazada)
compare = np.hstack((prueba1,img_prueba1_desplazada))
Image.fromarray(compare)

```

```python
"""## **PRUEBA 2**"""

#IMAGEN ORIGINAL Y EJECUCIÓN DE FUNCIONES
filename="Imagen5.bmp"
prueba1=Image.open(filename) #Abrimos la imagen 1
prueba1_arr = np.asarray(prueba1) #Convierte imagen a arreglo

"""### Original v. Contracción

*   Valor mínimo = 50
*   Valor máximo = 180
"""

prueba1_contraida = contraccion(prueba1_arr,180,50) #Contracción
img_prueba1_contraida = Image.fromarray(prueba1_contraida)
histograma(prueba1_arr)
histograma(prueba1_contraida)
compare = np.hstack((prueba1,img_prueba1_contraida))
Image.fromarray(compare)
```

```python
"""## **PRUEBA 3**"""

#IMAGEN ORIGINAL Y EJECUCIÓN DE FUNCIONES
filename="Imagen5.bmp"
prueba1=Image.open(filename) #Abrimos la imagen 1
prueba1_arr = np.asarray(prueba1) #Convierte imagen a arreglo


"""### Original v. Expansión

*   Valor mínimo = 190
*   Valor máximo = 30


"""

prueba1_expandida = expansion(prueba1_arr,200,0) #Expansión
img_prueba1_expandida = Image.fromarray(prueba1_expandida)
histograma(prueba1_arr)
histograma(prueba1_expandida)
compare = np.hstack((prueba1,img_prueba1_expandida))
Image.fromarray(compare)

```
